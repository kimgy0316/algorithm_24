###  2) 알고리즘 설명  (MD)

n이 짝수이면 x^n = (x^2)^n/2의 공식을 이용해 x^2을 먼저 계산한 후에 이 값을 n/2승한다. n이 홀수이면 x^2을 (n-1)/2승 하고 여기에 x를 곱해준다.

### 4) 코드 개요(입력 변수, 출력, 함수 설명) (MD)

##### 함수명: power(x,n)
#####         재규함수 power을 사용하여 입력값이 짝수이면 짝수를 검사하는 조건문에 (x*x,n//2)을 통과해
#####         출력시키고 입력값이 홀수이면 홀수 조건문을 검사하는 x*power (x*x,(n-1)//2)공식에 대입을 한다.
#####         반복문을 통해 2.0을 500번 제곱한 값을 반환하고 억지기법과 축소정복기법의 시간을 계산해 더 나은 #####         알고리즘과 알 수 있게 출력한다.
##### 입력값: 
##### power(2.0,500)
##### slow_power(2.0,500)
##### 반환값: 
##### print("억지기법(2**500)=",power(2.0,500))
##### print("축소정복기법(2**500)=",slow_power(2.0,500))

### 5) 코드 (PY,C)

```python
def slow_power(x,n):
    result = 1.0
    for i in range(n):
        result = result *x
    return result

def power(x,n):
    if n==0:
        return 1
    elif(n%2)==0:
        return power(x*x,n//2)
    else:
        return x*power(x*x,(n-1)//2)
    
import time
print("억지기법(2**500)=",power(2.0,500))
print("축소정복기법(2**500)=",slow_power(2.0,500))

t1 = time.time()
for i in range(100000): power(2.0,500)
t2 = time.time()
for i in range(100000): slow_power(2.0,500)
t3 = time.time()
print("억지기법 시간 : ",t3-t2)
print("축소정복기법 시간 :",t2-t1)
```
