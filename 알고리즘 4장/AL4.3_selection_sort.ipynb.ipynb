### 4) 코드 개요(입력 변수, 출력, 함수 설명) (MD)

#### 함수명 : insertion_sort
#### 입력 : data = [5, 3, 8, 4, 9, 1, 6, 2, 7]
#### 반환값 : 반환값은 따로 없고 정렬된 요소를 출력합니다

### 5) 코드 (PY,C)

```python
def insertion_sort(A):
    n = len(A)
    for i in range(1, n):
        key = A[i]
        j = i - 1
        while j >= 0 and A[j] > key:
            A[j + 1] = A[j]
            j = j - 1
        A[j + 1] = key
        printStep(A, i)
```

### 8) 복잡도 분석 (MD)

##### 리스트가 정렬되지 않았기 때문에 내부 루프가 i번 반복되므로 기본 연산이 i번 실행된다.
