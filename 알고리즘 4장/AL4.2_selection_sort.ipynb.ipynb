### 4) 코드 개요(입력 변수, 출력, 함수 설명) (MD)

##### 1) 함수명: def factorial_iter(n), 축소 정복 기법: 하향식 접근
##### 2) 입력 변수: 1, n+1
##### 3) 반환값: result

### 5) 코드 (PY,C)

```python
def factorial_iter(n) :
    result = 1
    for k in range(1, n+1) :
        result = result * k
    return result
```

### 6) 테스트 코드  (PY,C) , 입력을 바꾸어서

```python
def factorial_iter(n) :
    result = 1
    for k in range(1, n+1) :
        result = result * k
    return result

print(f"8!는 {factorial_iter(10)}")
```
