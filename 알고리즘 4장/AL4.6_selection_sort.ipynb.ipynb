### 1) 문제 정의 (MD)

리스트에 n개의 항목이 들어있다. 이 리스트에서 "탐색키"를 가진 항목을 찾아라. 단, 리스트의 항목들은 정렬되어 있다.

### 2) 알고리즘 설명  (MD)

이진 탐색 알고리즘은 반복구조나 순환구조로 모두 기술할 수 있다. 이진 탐색을 위한 실행 효율성을 위해서는 반복구조가 더 유리하다.

### 4) 코드 개요(입력 변수, 출력, 함수 설명) (MD)

##### 1) 함수명 : def binary_search_iter(A, key, low, hugh), 정렬된 리스트 A와 킷값 key, 그리고 리스트 내의 탐색 범위인 low와 high가 입력된다.
##### 2) 입력 변수: low, hugh
##### 3) 반환값 : 탐색에 성공하면 인텍스 mid를 반환하고, 탐색에 실패하면 -1을 반환한다.

### 5) 코드 (PY,C)

```python
def binary_search_iter(A, key, low, high) :
    while (low <= high) :
        mid = (low + high) // 2
        if key == A[mid]:
            return mid
        elif key > A[mid]:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

### 6) 테스트 코드  (PY,C) , 입력을 바꾸어서

### 8) 복잡도 분석 (MD)
