###  2) 알고리즘 설명  (MD)

주어진 X의 n 제곱을 계산하여 출력하는 코드로 이 코드에는 **연산자를 사용하지 않고 루프를 사용하여 각각의 값을 계산하는 억지기법이 사용된 알고리즘이다.

### 4) 코드 개요(입력 변수, 출력, 함수 설명) (MD)

##### 함수명 : slow_power
##### 입력 : x(거듭제곱의 밑이 되는 숫자),n(거듭제곱의 지수)
##### 반환값 : slow_power 함수에서 계산한 거듭제곱의 결과

### 5) 코드 (PY,C)

```python
def slow_power(x, n):
    result = 1.0
    for i in range(n):
        result = result * x
    return result
```

### 6) 테스트 코드  (PY,C) , 입력을 바꾸어서

```python
def slow_power(x, n):
    result = 1.0
    for i in range(n):
        result = result * x
    return result
x = 2
n = 10
print(slow_power(x, n))
```
