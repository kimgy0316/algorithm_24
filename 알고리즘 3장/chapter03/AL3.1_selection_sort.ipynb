### 5) 알고리즘 코드 (PY,C)

```python
def printstep(arr, val):
    print(" step %2d = " % val, end='')
    print(arr)

def selection(A):
    n=len(A)
    for i in range(n-1):
        least =i
        for j in range(i+1, n):
            if(A[j]<A[least]):
                least =j
        A[i], A[least] = A[least], A[i]
        printstep(A, i+1)

data = [5,3,8,4,9,1,6,2,7]
print("original", data)
selection(data)
print("selection : ",data)
```

### 6) 테스트 코드  (PY,C) , 입력을 바꾸어서

```python
def printstep(arr, val):
    print(" step %2d = " % val, end='')
    print(arr)

def selection(A):
    n=len(A)
    for i in range(n-1):
        least =i
        for j in range(i+1, n):
            if(A[j]<A[least]):
                least =j
        A[i], A[least] = A[least], A[i]
        printstep(A, i+1)

data = [7,1,8,4,10,1,16,2,17]
print("original", data)
selection(data)
print("selection : ",data)
```

### 8) 복잡도 분석

#### 알고리즘에서 이중 루프가 사용되는데, 외주 루프(3행)는 n-1번 반복하고, 내부 루프(5행)는 0에서 n-2까지 변하는 i에 대하여 (n-1)-i번 반복된다. 결국, 선택 정렬은 시간 복잡도가 O(n^2)이다.
