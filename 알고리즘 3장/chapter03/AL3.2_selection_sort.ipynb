### 4) 알고리즘 개요(입력 변수, 출력, 알고리즘 개요) (MD)

#### A=입력 리스트,key=탐색키
찾는 항목이 있으면 그 항목의 인덱스를 반환
찾지 모하면 -1를 반환

### 5) 알고리즘 코드 (PY,C)

```python
def sequential_search(A, key):
    for i in range(len(A)):
        if A[i] == key:
            return i
    return -1
```

### 6) 테스트 코드  (PY,C) , 입력을 바꾸어서

```python
def sequential_search(A, key):
    for i in range(len(A)):
        if A[i] == key:
            return i
    return -1
result = sequential_search([1,2,3,4,5], 1)
print(result)
```
