### 1) 문제 정의 (MD)

#### 리스트에 n개의 항목이 들어 있다. 이 리스트에서 "탐색키"를 가진 항목을 찾아라. 만약 찾는 항목이 있으면 그 항목의 인덱스를 반환하고, 없으면 -1을 반환하라. 단, 리스트의 항목들은 정렬되어 있지 않다.

### 2) 알고리즘 설명 (MD)

#### 입력 리스트의 첫 항목부터 순서대로 하나씩 탐색키와 비교한다.

### 3) 손으로 푼 예제 (MD, 이미지 삽입)

### 4) 알고리즘 개요(입력 변수, 출력, 알고리즘 개요) (MD)

#### 1행 : A는 입려 리스트,key는 탐색 키
#### 2행 : i:0.1, ..., len(A)-1
#### 3행 : 탐색 성공하면 (비교 연산, 기본 연산임)
#### 4행 : 인덱스 반환
#### 5행 : 탐색에 실패하면 -1 반환

### 5) 알고리즘 코드 (PY,C)

```python
def sequential_search(A, key):
    for i in range(len(A)):
        if A[i] == key:
            return i
    return -1
```

### 6) 테스트 코드 (PY,C) , 입력을 바꾸어서

```python
def sequential_search(A, key):
    for i in range(len(A)):
        if A[i] == key:
            return i
    return -1
result = sequential_search([1,2,3,4,5], 1)
print(result)
```

### 7) 수행 결과 (MD, 결과 캡춰하여 이미지로 삽입)

### 8) 복잡도 분석

#### 최선의 경우 : 리스트이 첫 번때 항목이 key인 경우로 한 번의 비교만으로 탐색이 종료된다. 따라서 시간 복잡도는 T(n)=1이고, O(n)이다.
#### 최악의 경우 : 찾고자 하는 숫자가 리스트에 없거나 맨 뒤에 있다면 리스트의 모든 항목을 검사해야 한다. 따라서 복잡도 함수는 T(n)=n이고, O(n)이다.
#### 평균적인 경우 : 리스트의 모든 숫자가 골고루 한 번씩 key로 사용되는 경우를 평균적인 경우라 가정하고 비교 횟수를 구하면 T(n)=(n+1)/2이고, O(n)이다.
