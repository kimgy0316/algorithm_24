#### 4) 알고리즘 개요(입력 변수, 출력, 알고리즘 개요) (MD)

##### 함수명 : def quick_sort(A, left, right) -> A[left, right]를 오름차순으로 정렬
##### 입력변수 : A, left, right
##### 반환값 : 없음

#### 5) 알고리즘 코드 (PY,C)

```python
def quick_sort(A, left, right):
    if left < right:
        pivot_index = partition(A, left, right)
        quick_sort(A, left, pivot_index - 1)
        quick_sort(A, pivot_index + 1, right)
```

#### 6) 테스트 코드  (PY,C) , 입력을 바꾸어서

```python
def quick_sort(A, left, right):
    if left < right:
        pivot_index = partition(A, left, right)
        quick_sort(A, left, pivot_index - 1)
        quick_sort(A, pivot_index + 1, right)

data = [5, 3, 8, 4, 9, 1, 6, 2, 7]
print("Original:", data)
quick_sort(data, 0, len(data) - 1)
print("QuickSort:", data)
```
